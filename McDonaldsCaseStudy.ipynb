```python
x = ds.iloc[:,0:11]
x
```

```python
x = (x == 'Yes').astype(int)
print(round(np.mean(x,axis=0),2))
```

```python
print(x)
```

```python
x.head()
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
  kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state=42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title('the elbow method')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(x)
```

```python
y_kmeans
```

```python
for i in range(0, len(y_kmeans)):
  print(y_kmeans[i])
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
new_ds = pca.fit_transform(x)
```

```python
new_ds
```

```python
kmeans.labels_
```

```python
plt.figure(figsize=(9,7))
plt.scatter(new_ds[:,0],new_ds[:,1],s=100, c = kmeans.labels_, cmap='coolwarm')
plt.xlabel("C1")
plt.ylabel("C2")
plt.title("Clusters")
plt.show()
```

```python
segment_names = {
    0: 'burger lovers',
    1: 'healthy eaters',
    2: 'fries lovers',
    3: 'chicken lovers'
}
```

```python
df = ds.apply(pd.to_numeric, errors = 'ignore')
df.describe()
```

```python
df['cluster'] = kmeans.labels_
df['segment'] = df['cluster'].map(segment_names)
```
